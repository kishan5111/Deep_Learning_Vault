```python
# Create a random array of numbers between 0 & 10 and an array of ones..both of size (3, 5), save them both to variables
random_array1 =  np.random.randint(0,10, size =(3,5))
ones1 = np.ones((3,5))
 
```

```python
# add both arrays
variable = np.add(random_array1 , ones1)
variable
```

```python
# add different shape arrays
ones2 = np.ones((5, 3))
new_var =ones2 + variable.T
```

```python
# create array of ones of shapes (3,5)
ones3 = np.ones((3,5))
```

```python
# substract ones array from latest
variable1 = np.subtract(new_var, ones3.T)
```

```python
# multiply the ones array with the latest array
variable2 = np.multiply(variable1, ones3.T)
```

```python
# take the latest array to the power of 2 using '**'
variable3 = variable2 ** 2
variable3
```

```python
variable4 = np.square(variable2)
variable4
```
